```python
#rename columns
sp_df = sp_df.rename(columns = {'Date': 'date', 'Open': 'open', 'High': 'high', 'Low': 'low', 
                                'Close': 'close', 'Volume': 'volume'})

#crate a new column with the % market change
column_values = sp_df['close']
pct_diff = column_values.pct_change() * 100
sp_df['percent_change'] = pct_diff

#realocate the % change column to the sixth spot
sp_df.insert(5, 'percent_change', sp_df.pop('percent_change'))

#crate a new column with the % volume change
column_values = sp_df['volume']
pct_diff = column_values.pct_change() * 100
sp_df['volume_percent_change'] = pct_diff

#eliminate rows with dates prior to 1928
sp_df = sp_df[sp_df['date'] >= '1928-01-01']
sp_df = sp_df.reset_index (drop=True)

#convert NaN to 0
sp_df['percent_change'] = sp_df['percent_change'].fillna(0)
sp_df['date'] = sp_df['date'].fillna(0)
sp_df['open'] = sp_df['open'].fillna(0)
sp_df['high'] = sp_df['high'].fillna(0)
sp_df['low'] = sp_df['low'].fillna(0)
sp_df['close'] = sp_df['close'].fillna(0)
sp_df['percent_change'] = sp_df['percent_change'].fillna(0)
sp_df['volume'] = sp_df['volume'].fillna(0)
sp_df['volume_percent_change'] = sp_df['volume_percent_change'].fillna(0)

# check the number of the rows contain inf values
inf_cols = sp_df.isin([np.inf, -np.inf]).any()
inf_rows = sp_df.isin([np.inf, -np.inf]).any(axis=1)
# print the columns and rows containing inf values
print("Rows containing inf: ", list(inf_rows[inf_rows == True].index))

#treat inf values
inf_df = sp_df.isin([np.inf, -np.inf])
inf_rows = sp_df.loc[inf_df.any(axis=1)]
print(inf_rows)
print(sp_df.iloc[6492])
sp_df = sp_df.replace([np.inf, -np.inf], 0)
```
