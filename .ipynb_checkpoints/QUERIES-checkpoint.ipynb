```python
import pandas as pd
```

```python
file = open(r'C:\Users\PC\Desktop\Ironhack\WR_Ironhack_Projects\Stock&Moon\sp500_index.csv')
sp_df = pd.read_csv(file)
sp_df
```

```python
crash_1929 = pd.DataFrame(sp_df.loc[(sp_df['date'] >= '1929-09-16') & (sp_df['date'] <= '1932-07-08')])

start_row = crash_1929.loc[crash_1929['date'] == '1929-09-16']
end_row = crash_1929.loc[crash_1929['date'] == '1932-07-08']

# calculate percentage difference between high and low values
percentage_diff = ((end_row['low'].values[0] - start_row['high'].values[0]) / start_row['high'].values[0]) * 100
percentage_diff

crash_1929['crash_1929'] = percentage_diff
```

```python
crash_1929
```

```python
crash_1929.to_csv(r'C:\Users\PC\Desktop\Ironhack\WR_Ironhack_Projects\Stock&Moon\crash_1929.csv', index = False)
```
